```python
import numpy as np
import matplotlib.pyplot as plt

# REPLACE these with your actual x and y values from manual work
x = np.array([1,3], dtype=float)
y = np.array([3,6], dtype=float)

n = len(x)
```

```python
m = -1   # initial slope (replace if different in your work)
b = 1   # initial intercept
alpha = 0.01  # learning rate (use the same you used manually)
iterations = 5
```

```python
m_values = []
b_values = []
errors = []

for i in range(iterations):  # Should be 'iterations', not 'iteration'
    
    # Step 1: Predictions
    y_pred = m * x + b
    
    # Step 2: Error (Mean Squared Error with factor 1/2n to match manual calculation)
    error = np.sum((y - y_pred) ** 2) / (2 * n)  # ✓ Now matches J = (1/2n)Σ(eᵢ²)
    
    # Step 3: Gradients (matching your manual derivation)
    dm = (-1/n) * np.sum(x * (y - y_pred))   # ✓ ∂J/∂m = -(1/n)Σ(eᵢ·xᵢ)
    db = (-1/n) * np.sum(y - y_pred)         # ✓ ∂J/∂b = -(1/n)Σ(eᵢ)
    
    # Step 4: Update parameters
    m = m - alpha * dm
    b = b - alpha * db
    
    # Store values
    m_values.append(m)
    b_values.append(b)
    errors.append(error)
    
    # Print each step clearly
    print(f"Iteration {i+1}")
    print(f"  m = {m:.6f}")
    print(f"  b = {b:.6f}")
    print(f"  Error = {error:.8f}")
    print("-------------------------")
```

```python
final_predictions = m * x + b

print("Final m:", m)
print("Final b:", b)
print("Final Predictions:", final_predictions)
```
